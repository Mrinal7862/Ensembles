```python
# Handeling the missing value
# data_encoding
# feature_scaling

from sklearn.impute import SimpleImputer # For missing Value
from sklearn.preprocessing import OneHotEncoder # for Encoding
from sklearn.preprocessing import StandardScaler # for feature scaling

from sklearn.pipeline import Pipeline # A sequence of Data transformers
from sklearn.compose import ColumnTransformer # Groups all the Pipeline  steps for each of transformers
```

```python
cat_cols = ['sex', 'smoker', 'day', 'time']
```

```python
num_cols = ['total_bill', 'tip', 'size']
```

```python
# Feature Engineering Automation using Pipe Line and Column Transformer

num_pipeline = Pipeline(steps=[
    ('imputation', SimpleImputer(strategy='median')),
    ('scaling', StandardScaler())
])

cat_pipeline = Pipeline(steps=[
    ('imputation', SimpleImputer(strategy='most_frequent')),
    ('encoding', OneHotEncoder())
])

```

```python
preprocessor = ColumnTransformer([('num_pipeline', num_pipeline, num_cols),
                   ("cat_pipeline", cat_pipeline, cat_cols)])
```

```python
preprocessor
```

```python
x_train = preprocessor.fit_transform(x_train)
x_test = preprocessor.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
model ={
    "support_vector_classifier": SVC(),
    "decision_tree": DecisionTreeClassifier(),
    "logistic_regression": LogisticRegression(),
    "random_forest": RandomForestClassifier()
}
```

```python
# Model train and Eval
from sklearn.metrics import accuracy_score
def model_train_eval(x_train, x_test, y_train, y_test, model):

  evaluation= {}

  for i in range(len(model)):

    models = list(model.values())[i]

    models.fit(x_train, y_train)

    y_pred = models.predict(x_test)

    accuracy = accuracy_score(y_test, y_pred)

    evaluation[list(model.keys())[i]] = accuracy

  return evaluation
```

```python
model_train_eval(x_train, x_test, y_train, y_test, model)
```

```python
# using Train test split

Xr_train, Xr_test, yr_train, yr_test = train_test_split(xr, yr, test_size=0.25, random_state=1)
```

```python
Xr_train.shape, Xr_test.shape
```

```python
Xr_train = preprocessor.fit_transform(Xr_train)
Xr_test = preprocessor.transform(Xr_test)
```

```python
# Handeling the missing value
# data_encoding
# feature_scaling

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
cat_colsr = ['sex', 'smoker', 'day', 'time']
num_colsr = ['tip', 'size']
```

```python
num_pipeline = Pipeline(steps=[
    ('imputation', SimpleImputer(strategy='median')),
    ('scaling', StandardScaler())
])

cat_pipeline = Pipeline(steps=[
    ('imputation', SimpleImputer(strategy='most_frequent')),
    ('encoding', OneHotEncoder())
])

preprocessor = ColumnTransformer([('num_pipeline', num_pipeline, num_colsr),
                   ("cat_pipeline", cat_pipeline, cat_colsr)])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor

modelr = {
    "linear_regression": LinearRegression(),
    "support_vector_regressor": SVR(),
    "random_forest": RandomForestRegressor()
}
```

```python
# Model train and Eval

from sklearn.metrics import r2_score
def model_train_eval_r(Xr_train, Xr_test, yr_train, yr_test, modelr):

  evaluationr= {}

  for i in range(len(modelr)):
    modelsr = list(modelr.values())[i]
    modelsr.fit(Xr_train, yr_train)
    y_predr = modelsr.predict(Xr_test)
    r2_scorer = r2_score(yr_test, y_predr)
    evaluationr[list(modelr.keys())[i]] = r2_scorer

  return evaluationr
```

```python
model_train_eval_r(Xr_train, Xr_test, yr_train, yr_test, modelr)
```

```python
# second approach
rf_r = RandomForestRegressor()

params ={
    'max_depth' : [1, 2, 3, 5, 10, None],
    'n_estimators': [100,50, 200, 1000],
    'criterion' : ['squared_error', 'friedman_mse', 'poisson', 'absolute_error']
}

reg = RandomizedSearchCV(rf_r, param_distributions=params, cv=5, verbose=3, scoring='r2')
```
